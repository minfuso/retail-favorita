The `holidays_events.csv` file contains contains calendar information about public holidays and special events on the period from 2012-03-02 to 2017-12-26. There is a total of 350 records.

There is 6 different holidays types which are 'Holiday', 'Transfer', 'Additional', 'Bridge', 'Work Day', 'Event'.
- 'Holiday' refers to an official public holiday
- 'Transfer' indicates that a official public holiday has been officially moved to this date.
- 'Additional' indicates an extra holiday
- 'Bridge' indicates a bridging day between a holiday and a weekend
- 'Work day' indicates a day that would normally be a holiday but has been designated as a working day.
- 'Event' indicates a day that would normally be a holiday but has been designated as a working day.

The variable **locale** indicates if the holiday is rather local to the city, the region, or is national. Therefore, the **locale_name** variable tells in which locality where the holiday applies.

The variable **description** is a textual description of the observed holiday.

Finally, the **transferred** boolean variable indicates whether a holiday has been moved to another date. When set to True, the original date is treated as a working day, while a corresponding entry with type 'Transfer' indicates the new observed holiday date.
